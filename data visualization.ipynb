```python
# Import required libraries here
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as s
```

```python
# Load dataset
data=pd.read_csv('C:/Users/jasro/DPhi.csv')
data.head()
```

```python
# Your code goes here
data=data.replace(to_replace='-',value='0')
data
```

```python
# Quiz score columns are categorical in nature, convert it to numerical
# Your code goes here
data['Quiz1']=data['Quiz1'].astype('float64')
data['Quiz2']=data['Quiz2'].astype('float64')
data['Quiz3']=data['Quiz3'].astype('float64')
data['Quiz4']=data['Quiz4'].astype('float64')
data['Quiz5']=data['Quiz5'].astype('float64')
data['Quiz6']=data['Quiz6'].astype('float64')
data['Quiz7']=data['Quiz7'].astype('float64')
```

```python
# Your code goes here
sns.countplot(x='Learner_Category',hue='Learner_Category',data=data)
data['Learner_Category'].value_counts()
```

***Write your inferences here:***



1.  Absolute Beginner = 635 
2.  Beginner = 812 
3. number of beginner> number of absolute beginner




```python
# Your code goes here
data['Total_Score']=data['Total_Score'].astype('float64')
total_score_beginner=data[(data['Total_Score']>0)&(data['Learner_Category']=='Beginner')]
plt.hist(total_score_beginner.Total_Score)
plt.xlabel('Total Score')
plt.ylabel('Frequency')
plt.title('Distribution of Total Scores of Beginners')
plt.show()
```

```python
# Your code goes here
total_score_abs_beginner=data[(data['Total_Score']>0)&(data['Learner_Category']=='Absolute Beginner')]
plt.hist(total_score_abs_beginner.Total_Score)
plt.xlabel('Total Score')
plt.ylabel('Frequency')
plt.title('Distribution of Total Scores of Absolute Beginners')
plt.show()
```

***Write your inferences here:***


1. Majority of absolute beginners scores lies between 50 to 64 aprox., but the distribution ranges from 4 to 68 approx.     
2. The minority or minimum absolute beginners scored marks from 24 to 30 approx.  
3. In case of beginners the majority of total score lies between 50 to 55 aprox., while the minority lies between 30 to 35     
   approx.
4. There is also a significant difference in frequency of distribution of beginners and absolute beginners as 
   number of beginners> number of absolute beginners. 

```python
grouped=total_score_abs_beginner.groupby('Group_ID')
mean_table=grouped.aggregate(np.mean)
mean_table
```

```python
mean_table.index
```

```python
a= plt.figure()
axes= a.add_axes([1.5,1.5,1.5,1.5])
plt.plot(mean_table.index.tolist(),mean_table['Quiz1'].tolist(),linestyle='--', marker='x',label='quiz1')
plt.plot(mean_table.index.tolist(),mean_table['Quiz2'].tolist(),linestyle='--', marker='x',color='red',label='quiz2')
plt.plot(mean_table.index.tolist(),mean_table['Quiz3'].tolist(),linestyle='--', marker='x',color='magenta',label='quiz3')
plt.plot(mean_table.index.tolist(),mean_table['Quiz4'].tolist(),linestyle='--', marker='x',color='green',label='quiz4')
plt.plot(mean_table.index.tolist(),mean_table['Quiz5'].tolist(),linestyle='--', marker='x',color='black',label='quiz5')
plt.plot(mean_table.index.tolist(),mean_table['Quiz6'].tolist(),linestyle='--', marker='x',color='yellow',label='quiz6')
plt.plot(mean_table.index.tolist(),mean_table['Quiz7'].tolist(),linestyle='--', marker='x',color='grey',label='quiz7')
plt.xlabel('Groups',size='15')
plt.ylabel('Mean of Quiz Scores',size='15')
plt.xticks(rotation=45)
plt.title("Mean Score trends of absolute beginners",size='15')
plt.legend(loc='best',bbox_to_anchor=(1.115, 1))
plt.show()
```

***Write your inference here:***


1. The average score in quizzes of absolute begginers decrease as per quiz number increases (ex- quiz 7 has the maximum average of 3 approx. , while of other quizzes ares still comparable)
2. The graph is not so smooth representing different people capabilities due to diversity of absolute begginers.
